```python
import numpy as np
import matplotlib.pyplot as plt

# cargar las librerias
from sklearn import datasets
```

```python
# funci칩n linspace

x = np.linspace(-10,10,100)

y = 2*x

#plt.plot(x,y)

# valores limites de los ejes
plt.xlim(0,10)
plt.ylim(0,10)

# u=unico, en python 3 no es necesario
plt.title(u"Gr치ficar de una funci칩n lineal\n")

# graficas de lineas discontinuas y punteadas
plt.plot(x,y, 'm*')

# graficar
plt.show()

```

```python
# funciones 
y = np.sin(x)
y1 = np.cos(x)
y3 = np.tan(x)

# dos graficas en un solo plano
plt.plot(y, 'c.',y1,'y--')
plt.show()


plt.plot(y,'m.', label='seno')
plt.show()


plt.plot(y3, 'r.',label = 'tangente')
# valores limites de los ejes
plt.xlim(-180,180)
plt.ylim(-5,5)
plt.show()
print(np.tan(2*np.pi))
```

```python
# Distribuciones Gausianas
N = 10

# Para generar valores aleatorios bajo la normal
valores = np.random.randn(N)
print(valores)
#plt.scatter(valores,c='r')
#plt.show()

mu,sigma =0,2
x =np.linspace(-3,3,100)
p=(1/(np.sqrt(2*np.pi)*sigma))*np.exp((-1/2)*(((x-mu)**2)((sigma))))

plt.plot(x,p,'m*')
plt.show()
```
