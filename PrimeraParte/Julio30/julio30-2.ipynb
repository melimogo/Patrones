## Metodologias de validación

¿Qué es una metodologia de validación?
    
    * Es una estrategia para que se particione una bd, Para tener un modelo de entrenamiento y otro de validación
        - Entrenamiento: Aprender unos parametros
        - Validación: Garantizar que cuando llegue una nueva muestra la clasifico bn
        - testing data: clientes nuevos
        
### Validación cruzada

Dividir la bd en k subconjuntos aleatorios disyuntos, usar k-1 para entrenar y el 1er k para validar

k iteraciones == k folds 


### Validación cruzada(leave-one-out)
    
        Solo valido con una bd muy pequeña

        entreno con k-1 y valido con la ultima, se hace nxn veces
        
## Validación Booststrap

    Se particiona la bd asi: 
        - train=70%
        - test = 30% 
        
     Se debe de hacer varias veces y de forma aleatoria las particiones.
     
     En un bd muy grande se puede hacer 50-50
     
     ¿Cuantas veces se repita?
           Deprende del numero de muestras que se tenga. 
           
          booststraping de forma regular se hace de a 5
      
      
### Base de datos desbalanceada

    Muchas muestras de una clase y pocas de otra
    
    Hay que garantizar que las particiones se tomen de manera proporcional 70% de cada una de las clases
